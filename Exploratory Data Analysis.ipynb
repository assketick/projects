Используйте для линий разные цвета, одна из них пусть будет пунктирной. Результат в этом и следующих заданиях на построение графиков должен получиться примерно как на картинке.
```python
fig, ax = plt.subplots()
x1, x2 = np.linspace(-np.pi, np.pi, 100), np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x1)
y2 = np.cos(x2)
ax.plot(x1, y1, 
        label='y=sin(x)', 
        color = '#18F7FF')
ax.plot(x2, y2, 
        label = 'y=cos(x)',
        color = '#EC57FF',
        linestyle = '--')
ax.legend()
ax.grid()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = data.drop('credit_risk', 1).values
y = data['credit_risk'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
def normalization(X: np.array) -> np.array:
    return np.abs(X - X.mean(axis = 0)) / X.std(axis = 0)

def predict(Xtest: np.array) -> np.array:
    weights = np.array([0.57, 0.01, 0.01, 0.12, 0.21, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01])
    return np.around(Xtest @ weights)

X_train = normalization(X_train)
X_test = normalization(X_test)

accuracy_score(y_test, predict(X_test))
```
